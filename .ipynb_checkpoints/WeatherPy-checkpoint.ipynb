## Observations

1. From the scatter plots, we can see that as you approach the Equator (0˚ latitude) the maximum city temperate tends to increase. This is especially evident when looking at the Northern Hemisphere vs. Maximum Temperature plots, which shows a strong linear relationship between latitude and maximum temperature with an r-squared value of .7, signifying much of the variability in temperature is accounted for by change in city latitude.
2. There does not seem to be a strong linear relationship between latitude and cloudiness percentage or latitude and humidity percentage; further, there is a wide range of values for both cloudiness and humidity percentages overall and in both hemispheres.
3. Finally, there does not seem to be a strong linear relationship between latitude and wind speed, but when we examine the Northern and Southern hemisphere plots we do see some differences between the hemispheres: The range of wind speeds in the Southern hemisphere is smaller than the range in the Northern Hemisphere. The wind speeds tend to stay between 0 and 15 mph in the southern hemisphere, but in the Northern hemisphere the range doubles include a maximum of about 30 mph. However, it is also important to note that 68% of the cities in the dataset fall in the Northern hemisphere so we have more observations in the North.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (.csv)
output_data_file = "Data/cities.csv"    

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# url with API key
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

# initialize lists for each column to append to a df
city_name = []
country = []
date = []
latitude = []
longitude = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []

# create a city counter
city_count = 0

# create a set counter that begins with set 1 to group the records
set_count = 1

# loop through all the cities in our list
for x, city in enumerate(cities):
    
    # logging and grouping
    if (x % 100 == 0 and x >= 100):
        set_count += 1     # move to next set
        city_count = 0     # back to 0
    
    # increment city count
    city_count += 1    
    
    # print log for city and city #
    print(f'{city}: city {city_count} in set {set_count}')
    
    # get data for each city
    try:    
        # url for each city
        query_url = url + "&q=" + city
        # perform data request for each city
        weather_json = requests.get(query_url).json()  
        
        # append data for each city to its respective list 
        city_name.append(weather_json["name"])
        country.append(weather_json["sys"]["country"])
        date.append(weather_json["dt"])
        latitude.append(weather_json["coord"]["lat"])
        longitude.append(weather_json["coord"]["lon"])
        max_temp.append(weather_json["main"]["temp_max"])
        humidity.append(weather_json["main"]["humidity"])
        cloudiness.append(weather_json["clouds"]["all"])
        wind_speed.append(weather_json["wind"]["speed"])
        
        # delay, if needed
        #time.sleep(.5)
        
    # if city is not avialible, continue to next city in cities
    except:
        print(f'{city} not found. Skipping!')
    continue
    
# indicate when done
print('------------------------------------')
print('Done!')
```

```python
# create weather df
weather_df = pd.DataFrame({"City": city_name,
                           "Date": date,
                           "Country": country,
                           "Lat": latitude, 
                           "Lng": longitude,
                           "Max Temp": max_temp,
                           "Humidity": humidity,
                           "Cloudiness": cloudiness,
                           "Wind Speed": wind_speed})

weather_df.head()
```

```python
# confirm we have enough observations
weather_df['City'].count()
```

### Convert Raw Data to DataFrame

```python
# save df to .csv file
weather_df.to_csv(output_data_file)

weather_df.head()
```

## Linear Regression

```python
# Create a function to create Linear Regression plots

def lin_regress_plot(x, y, title, xlab, ylab, eq_loc, file_name):

    # regression model
    regression = linregress(x, y)
    
    # regression equation to create line (y = mx + b)
    slope = regression[0]
    intercept = regression[1]
    line_to_plot = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    # report r-squared and p-val
    rsq_value = round(regression[2]**2,2)
    p_value = round(regression[3],3)
    
    # scatter plot with regression line
    plt.scatter(x, y, color = 'b', alpha = .5)
    plt.plot(x, line_to_plot, 'r-')
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.annotate(line_eq, eq_loc, fontsize = 14, color = 'r')
    plt.grid(True)
    print(f"The r-squared value is: {rsq_value}")
    print(f"The p-value  is: {p_value}")
    plt.savefig(file_name)
```

```python
# Create Northern and Southern Hemisphere DataFrames

# Northern (above Equator)
northern_df = weather_df[weather_df['Lat'] >= 0]

# Southern (below Equator)
southern_df = weather_df[weather_df['Lat'] < 0]

print(len(northern_df)/weather_df['City'].count() * 100)  # more cities in the Northern hemisphere
print(len(southern_df)/weather_df['City'].count() * 100)
```
